# Suma con lista tradicional
- Crea una lista de 100 millones de números (del 0 al 99,999,999).
- Usa un bucle for para sumar 1 en cada iteración (no suma los valores de la lista, solo cuenta).
- Mide cuánto tiempo tarda en ejecutarse ese bucle.
- Imprime el tiempo total de ejecución.
Objetivo: mostrar lo lento que puede ser un bucle tradicional con listas grandes.


```
lista = list(range(100_000_000))

inicio = time.perf_counter()

suma = 0
for i in lista:
  suma += 1

final = time.perf_counter()

tiempo = final - inicio

print(f"Tiempo de ejecución:{tiempo:6f} sec")
```

# Suma con NumPy
- Crea un arreglo de NumPy con los mismos 100 millones de números.
- Usa np.sum() para sumar todos los valores del arreglo.
- Mide el tiempo de ejecución.
- Imprime cuánto tardó.
Objetivo: comparar la eficiencia de NumPy vs. bucle tradicional. NumPy es mucho más rápido porque usa operaciones vectorizadas.


```
np_lista = np.arange(100_000_000)

inicio = time.perf_counter()

suma_np = np.sum(np_lista)

final = time.perf_counter()

tiempo = final - inicio

print(f"Tiempo de ejecución:{tiempo:6f} sec")
```

# Base de datos como lista de listas
- Define una base de datos como lista de listas (cada persona es una sublista).
- Usa dos diccionarios:
- columnas: para acceder por nombre de campo (ej. "Mascota" → índice 2).
- filas: para acceder por nombre de persona (ej. "Kevin" → índice 3).
- Imprime el animal favorito de Kevin: "Periquitos".
Objetivo: mostrar cómo acceder a datos específicos usando índices con ayuda de diccionarios.


```

base_de_datos = [
    ["Oscar", 21, "Gatos"],
    ["Paula", 21, "Tiburones"],
    ["Jonatan", 24, "Hurones"],
    ["Kevin", 22, "Periquitos"],
    ["Lucero", 21, "Perros"],
    ["Israel", 21, "Lobos"],
    ["Sergio", 21, "Perros"],
    ["Yahir", 21, "Leones"],
    ["Diego", 21, "Pingüinos"]
]
columnas = {"Nombre":0, "Edad":1, "Mascota":2}
filas = {"Oscar":0,"Paula":1,"Jonatan":2,"Kevin":3,"Lucero":4,"Israel":5,"Sergio":6,"Yahir":7}
print(base_de_datos[filas["Kevin"]][columnas["Mascota"]])
```

# Uso de pandas para visualizar y filtrar
- Convierte la lista base_de_datos en un DataFrame de pandas, con nombres de columnas y un índice del 1 al 9.
- Imprime toda la tabla.
- Filtra la columna Animal donde el nombre sea "Kevin" → muestra "Periquitos".
Objetivo: mostrar cómo pandas facilita la visualización y filtrado de datos de forma más clara y poderosa que listas tradicionales.


```
dF = pd.DataFrame(base_de_datos, columns=["Nombre", "Edad", "Animal"], index=range(1,10))
print(dF)
print(dF.Animal[dF['Nombre'] == 'Kevin'])
```

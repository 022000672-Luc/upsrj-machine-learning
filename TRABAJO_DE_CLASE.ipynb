```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

#Definición de los datos

Se establece el modelo matemático mediante la ecuación y = m·x + b, la cual describe una relación lineal entre las variables.
Se construye el arreglo X correspondiente a la variable independiente y se ajusta su forma a una estructura de columna, de modo que pueda utilizarse correctamente en el proceso de regresión lineal.
Posteriormente, se define el arreglo y como la variable dependiente del modelo.
Por último, se muestra en pantalla el contenido de y con el propósito de comprobar que los datos hayan sido cargados de manera correcta.


```python
# y = m * x + b

# Variable independiente
# SciKit-Learn requiere que se usen arreglos de 2D
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
print(x)

# Variable dependiente
y = np.array([2, 4, 5, 4, 5])
print(y)
```

#Graficación

Se elabora una gráfica de tipo dispersión a partir de los valores contenidos en X y y, con el fin de visualizar su relación.
El eje horizontal se identifica con la variable X, mientras que el eje vertical corresponde a la variable Y.
Se agrega un título que permite reconocer la gráfica como la representación de los datos originales.
Además, se habilita la cuadrícula para mejorar la lectura y el análisis de la información.
Finalmente, la gráfica es presentada en pantalla.
- y_pred: es un arreglo con los valores de y que tu modelo predice para cada x.
- Si no lo defines, Python no sabe qué graficar y lanza el NameError.
- Puedes calcularlo con tu modelo (ej. model.predict(x)) o inventar una recta provisional como arriba, solo para que se muestre en la gráfica


```python
# Generamos nuestro modelo de regrecion linela
model = LinearRegression()

# Alimentamos nuestro modelo
model.fit(x, y)

# Parametros:m(peso), b(sesgo)
m = model.coef_[0]
b = model.intercept_


print(f"Pendiente (m): {m:.2f}")
print(f"intercepto (b): {b:.2f}")
```

```python
from sklearn.linear_model import LinearRegression


model = LinearRegression()
model.fit(x.reshape(-1, 1), y)

y_pred = model.predict(x.reshape(-1, 1))

plt.scatter(x, y, label="Datos reales")
plt.plot(x, y_pred, color="red", label="Modelo lineal")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Datos originales")
plt.grid(visible=True)
plt.legend()
plt.show()
```

# Implementación de Pipeline

Un pipeline encapsula todo el flujo de procesamientos de datos en un solo objeto.

Para el usuario, entrenar un pipeline es igual que entrenar un modelo

El escalado no cambia el resultado conceptual, pero prepara el terreno para modelos más complejos.

## Justificación

Hasta ahora hemos entrenado modelos paso por paso. En la vida real, esos pasos siempre se repiten en el mismo orden. Un pipeline es una forma de decirle a la computadora: **"Estos son los pasos, siempre ejecútalos así"**

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
def plot_model(xlabel: str, ylabel: str, title: str):
  """
  Esta funcion gráfica un modelo de regrecion lineal.
  """
  plt.scatter(x, y, label="Datos reales")
  plt.plot(x, y_pred, color="red", label="Modelo lineal")
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  plt.title(title)
  plt.grid(visible=True)
  plt.legend()
  plt.show()
```

```python
#Definimos las variables dependientes e independientes de nuestro modelo.

#Variables independientes de x:Distancia al centro
x = np.arange(1, 21).reshape(-1, 1)
print(x)

#Variables dependientes
y = np.array([340, 320, 290, 260, 230,
              210, 220, 245, 275, 305,
              330, 350, 365, 375, 380,
              385, 390, 395, 398, 400])
print(y)
```

```python
# Definicion del pipline de regresion linela

pipline = Pipeline(steps=[
    ("scalar", StandardScaler()),
    ("regresor", LinearRegression())
])

# Entrenamiento del modelo
pipline.fit(x, y)

# Pedicion del modelo
y_pred = pipline.predict(x)

plot_model(
    xlabel="Distancia al centro (Km)",
    ylabel="Precio (miles de $)",
    title="Estimación de precio de una casa (regresión lineal simple)"
)
```
